```python
import numpy as np
from scipy.special import sph_harm, spherical_jn, spherical_yn, sph_harm_y
import matplotlib.pyplot as plt
```

# $X_{lm}$

En la ecuación 9.119 del capítulo 9 de Jackson se define el armónico vectorial $\vec{X}_{l,m}$ como:

\begin{equation}
    \vec{X}_{l,m} = \frac{\vec{L}Y_{l,m}}{\sqrt{l(l+1)}}.
\end{equation}

Donde $Y_{l,m}$ es el armónico esférico 

\begin{equation}
    Y_{l,m}(\theta, \varphi) = \sqrt{\frac{2l+1}{4\pi}\frac{(l-m)!}{(l+m)!}}P_l^m(\cos{\theta})e^{im\varphi},
\end{equation}

y el operador $\vec{L}$ es, según la ecuación 9.101 del mismo capítulo:

\begin{equation}
    \vec{L} = -i\vec{r}\times\nabla = -i\left(-\frac{1}{\sin{\theta}}\frac{\partial}{\partial \varphi}\hat{\theta}+\frac{\partial}{\partial \theta}\hat{\varphi}\right).
\end{equation}

Si aplicamos el operador al armónico $Y_{l,m}$:

\begin{equation}
    \vec{L}Y_{l,m}=-i\vec{r}\times \nabla Y_{l,m}=\left(\frac{i}{\sin{\theta}}\frac{\partial Y_{l,m}}{\partial\varphi}\hat{\theta}-i\frac{\partial Y_{l,m}}{\partial \theta}\hat{\varphi}\right)=-\frac{m}{\sin{\theta}}Y_{l,m}\hat{\theta}-i\frac{\partial Y_{l,m}}{\partial \theta}\hat{\varphi}
\end{equation}

por tanto:

\begin{equation}
    \textbf{X}_{l,m} = \frac{1}{\sqrt{l(l+1)}}\left(-\frac{m}{\sin{\theta}}Y_{l,m}\hat{\theta}-i\frac{\partial Y_{l,m}}{\partial \theta}\hat{\varphi}\right)
\end{equation}

La derivada de $Y_{l,m}$ con respecto a $\theta$ puede calcularse a través de los operadores $L_\pm$:

\begin{equation}
    L_{\pm} = e^{\pm i \varphi}\left( \pm \frac{\partial}{\partial \theta} + i \,cotg\,\theta\,\frac{\partial}{\partial \varphi}  \right)
\end{equation}

Que actúan sobre $Y_{l,m}$ como:

\begin{equation}
    L_{\pm}Y_{l,m} = \sqrt{(l \mp m )(l\pm m +1)}Y_{l,m\pm1}
\end{equation}

Sustituyendo y utilizando que $\frac{\partial}{\partial \varphi} Y_{l,m} = im Y_{l,m}$:

\begin{equation}
    e^{\pm i\varphi}(\pm \frac{\partial Y_{l,m}}{\partial \theta}  - m\,cotg\,\theta\,Y_{l,m})=\sqrt{(l\mp m )(l\pm m +1)}Y_{l,m\pm1}
    \end{equation}

Despejando:

\begin{equation}
        \frac{\partial Y_{l,m}}{\partial \theta}  = m\,cotg\,\theta Y_{l,m} +  e^{-i\varphi}\sqrt{(l-m)(l+m+1)}Y_{l,m+1}     
\end{equation}

\begin{equation}
        -\frac{\partial Y_{l,m}}{\partial \theta} = m\,cotg\,\theta Y_{l,m} -  e^{i\varphi}\sqrt{(l+m)(l-m+1)}Y_{l,m-1}     
\end{equation}

Restando estas ecuaciones se obtiene:

\begin{equation}
    \frac{\partial Y_{l,m}}{\partial \theta}  = \frac{1}{2}\left[ e^{-i\varphi}\sqrt{(l-m)(l+m+1)}Y_{l,m+1} - e^{i\varphi}\sqrt{(l+m)(l-m+1)}Y_{l,m-1}\right]
\end{equation}

Los armónicos esféricos $Y_{l,m}$ están directamente implementados en python a través del paquete scipy.special, la siguiente función es su derivada con respecto a $\theta$.




```python
def dY_theta(l, m, Theta, Phi): 
    Y_p = sph_harm_y(l, m+1, Theta, Phi)
    Y_n = sph_harm_y(l, m-1, Theta, Phi)

    
    c_p = np.sqrt((l - m) * (l + m + 1))
    c_n = np.sqrt((l + m) * (l - m + 1))
   
    dY = 0.5 * (
        c_p * Y_p * np.exp(-1j * Phi) 
        - c_n * Y_n * np.exp(1j * Phi)
    )
  
    return dY
```

Con esto podemos construir el armónico vectorial $X_{l,m}$:

```python
def X_lm(l, m, Theta, Phi):
    Y = sph_harm_y(l, m, Theta, Phi)
    dY_t = dY_theta(l, m, Theta, Phi)

    pre = 1/np.sqrt(l*(l+1))
    
    X_theta = pre * (1*m/np.sin(Theta)) * Y # ORIGINAL MIllan
    X_theta = pre * (-1*m/np.sin(Theta)) * Y # Corregido
    X_phi   = pre * -1j * dY_t

    return X_theta, X_phi
```

# $M_{l,m}$ y $N_{l,m}$

Los multipolos $M_{l,m}$ y $N_{l,m}$ se construyen a partir de $X_{l,m}$ de la siguiente forma (Sección 2A de Neves):

\begin{equation}
    \vec{M}_{l,m} = z_l(kr)\vec{X}_{l,m}
\end{equation}

\begin{equation}
    \vec{N}_{l,m} = \frac{i}{k}\nabla \times \vec{M}_{l,m}
\end{equation}

Donde $z_l(kr)$ corresponde a una función esférica de Bessel para campos incidentes y de Hankel de primera especie para campos dispersados, $k=2\pi/\lambda$ y $\lambda$ es la longitud de onda.

Podemos implementar $z_l(kr)$ y por tanto $\vec{M}_{l,m}$ de la siguiente forma, para calcular $\vec{N}_{l,m}$ necesitaremos también la derivada de $z_l(kr)$:

```python
def z_l(l, kr, forma='j'):
    if forma == 'j':
        return spherical_jn(l, kr+0j)
    if forma == 'h':
        return spherical_jn(l, kr) + 1j * spherical_yn(l, kr)
    
def dz_l(l, kr, forma='j'):
    if forma == 'j':
        return spherical_jn(l, kr+0j, True)
    if forma == 'h':
        return spherical_jn(l, kr, True) + 1j * spherical_yn(l, kr, True)

def M_lm(l, m, k, R, Theta, Phi, forma='j'):
    kr = k*R
    z = z_l(l, kr, forma)

    X_theta, X_phi = X_lm(l, m, Theta, Phi)
    
    nr = R.shape[0]
    ntheta = Theta.shape[1]
    nphi = Phi.shape[2]

    M_r     = np.zeros((nr, ntheta, nphi), dtype=np.complex64)
    M_theta = z * X_theta
    M_phi   = z * X_phi

    return M_r, M_theta, M_phi
```

Para obtener $\vec{N_{l,m}}$ calcularemos el rotacional de $\vec{M_{l,m}}$:

$$
\vec{N}_{l,m}=\frac{i}{k}\frac{1}{r^2\sin{\theta}}
\begin{vmatrix}
\hat{r} & r\hat{\theta} & r\sin{\theta}\hat{\varphi} \\
\frac{\partial}{\partial r} & \frac{\partial}{\partial \theta} & \frac{\partial}{\partial \varphi} \\
0 & rM_\theta & r\sin{\theta}M_\varphi 
\end{vmatrix}  
$$

Para la componente en r podemos utilizar la ecuación 31 de Neves:

\begin{equation}
    \hat{r}\cdot\vec{N}_{l,m} = N_r = -\frac{z_l(kr)}{kr}\sqrt{l(l+1)}Y_{l,m}
\end{equation}

Las componentes en $\theta$ y $\varphi$ se pueden obtener del determinante:

\begin{equation}
    \hat{\theta}\cdot\vec{N}_{l,m} = N_\theta = -\frac{i}{k}\frac{r\sin{\theta}}{r^2\sin{\theta}} \left( \frac{\partial}{\partial r}(rM\varphi)  \right) = -\frac{i}{kr}\left(M_\varphi+r\frac{\partial}{\partial r}M_\varphi\right) = -\frac{i}{kr}X_\varphi\left(z_l+kr\frac{d}{dr}z_l\right)
\end{equation}

\begin{equation}
    \hat{\varphi}\cdot\vec{N}_{l,m} = N_\varphi = \frac{i}{k}\frac{r\sin{\theta}}{r^2\sin{\theta}} \left( \frac{\partial}{\partial r}(rM_\theta)  \right) = \frac{i}{kr}\left(M_\theta+r\frac{\partial}{\partial r}M_\theta\right) = \frac{i}{kr}X_\theta\left(z_l+kr\frac{d}{dr}z_l\right)
\end{equation}

Implementamos también $\vec{N_{l,m}}$:

```python
def N_lm(l, m, k, R, Theta, Phi, forma='j'):
    kr = k*R
    z  = z_l(l, kr, forma)
    dz = dz_l(l, kr, forma)

    Y = sph_harm_y(l, m, Theta, Phi)

    X_theta, X_phi = X_lm(l, m, Theta, Phi)
 
    N_r =  z * 1/kr * np.sqrt(l*(l+1)) * Y # ORIGINAL Millan
    N_r =  -z * 1/kr * np.sqrt(l*(l+1)) * Y # Corregido
    N_theta = -1j/kr * X_phi * (z + kr * dz)
    N_phi   = 1j/kr * X_theta * (z + kr * dz)

    return N_r, N_theta, N_phi
```

# Beam Scattering Coefficients (BSCs)

En Neves 2A se explica cómo podemos reconstruir un campo electromagnético libre multiplicando los multipolos $M_{l,m}$ y $N_{l,m}$ por una serie de coeficientes.

\begin{equation}
    \vec{E}_{inc} = E_0\sum^\inf_{l=1}\sum^l_{m=-l}(G_{lm}^{TE}\vec{M}_{lm} +  G_{lm}^{TM}\vec{N}_{lm})
\end{equation}
\begin{equation}
    Z\vec{H}_{inc} = E_0\sum^\inf_{l=1}\sum^l_{m=-l}(G_{lm}^{TM}\vec{M}_{lm} -  G_{lm}^{TE}\vec{N}_{lm})
\end{equation}

Para una onda plana solo sobreviven los coeficientes en los que $m=\pm1$. Neves da las expresiones de estos coeficientes en el caso de una onda plana linealmente polarizada en x que se propaga a lo largo del eje z, las ecuaciones 17 y 18 del paper "Photonic nanojets in optical tweezers" (https://www.sciencedirect.com/science/article/abs/pii/S0022407315001119?via%3Dihub) nos da una versión más general de estos coeficientes:

ORIGINALES NEVES
\begin{equation}
    G_{l,1}^{TE} = G_{l,pw}(p_x-ip_y)
\end{equation}
\begin{equation}
    G_{l,-1}^{TE} = G_{l,pw}(p_x+ip_y)
\end{equation}
\begin{equation}
    G_{l,1}^{TM} = G_{l,pw}(-ip_x-p_y)
\end{equation}
\begin{equation}
    G_{l,-1}^{TM} = G_{l,pw}(ip_x-p_y)
\end{equation}

MODIFICADOS MILLAN
\begin{equation}
    G_{l,1}^{TE} = G_{l,pw}(ip_x-p_y)
\end{equation}
\begin{equation}
    G_{l,-1}^{TE} = G_{l,pw}(-ip_x-p_y)
\end{equation}
\begin{equation}
    G_{l,1}^{TM} = G_{l,pw}(p_x-ip_y)
\end{equation}
\begin{equation}
    G_{l,-1}^{TM} = G_{l,pw}(p_x+ip_y)
\end{equation}

donde:

\begin{equation}
    G_{l,pw}=i^l\sqrt{\pi(2l+1)}
\end{equation}

NOTA: Los coeficientes TE difieren de la referencia en un signo -1, aún no tengo claro por qué.

Podemos implementar los coeficientes para una onda plana de polarización dada por $p_x$ y $p_y$ de la siguiente forma:


```python
def BSC_plana(l):
    common = np.sqrt(np.pi*(2*l+1)) * 1j**(l) 
    px = 1
    py = 0
    # Modificados Millán
    G_TM_p = common * (+1j*px-py) 
    G_TM_m = common * (-1j*px-py) 
    G_TE_p = common * (px + 1j*py)
    G_TE_m = common * (px - 1j*py)
    # Originales Neves
    G_TM_p = common * (-1j*px-py) 
    G_TM_m = common * (+1j*px-py) 
    G_TE_p = common * (px - 1j*py)
    G_TE_m = common * (px + 1j*py)
    return G_TM_p, G_TM_m, G_TE_p, G_TE_m
```

# Campos en XZ (Parte Real)

Pasamos ahora al cálculo de los campos E y H, nos restringiremos al plano XY para agilizar la ejecución del código, inicializamos las variables.

```python
wl = 1.0                # longitud de onda
k = 2*np.pi/wl          # número de onda
z_imp = 1.0             # impedancia del medio
E0 = 1.0                # amplitud del campo
l_max = 70               # orden máximo del desarrollo
nr = 150               # puntos en r
nphi = 1             # puntos en phi
ntheta = 181              # Solo 1 punto en theta
r = np.linspace(1e-6, 10*wl, nr)
theta = np.linspace(1e-6, np.pi, ntheta) 
phi   = np.linspace(1e-6, 1e-6, nphi) 

R = r[:, None, None]         # forma (nr,1,1)
Theta = theta[None, :, None] # forma (1,1,1)
Phi = phi[None, None, :]     # forma (1,1,nphi)
```

Definimos también los campos en coordenadas esféricas:

```python
Er     = np.zeros((nr, ntheta, nphi), dtype=np.complex64)
Etheta = np.zeros_like(Er)
Ephi   = np.zeros_like(Er)

Hr     = np.zeros((nr, ntheta, nphi), dtype=np.complex64)
Htheta = np.zeros_like(Hr)
Hphi   = np.zeros_like(Hr)
```

Calculamos los distintos órdenes del desarrollo y vamos sumando en cada componente del campo:

```python
for l in range(1, l_max+1): 
    if l % 10 == 0: print(f"Procesando orden l={l}")
#    print(f"Procesando orden l={l}")
    
    G_TM_p, G_TM_m, G_TE_p, G_TE_m = BSC_plana(l)

    Nr_p, Nt_p, Np_p = N_lm(l, 1, k, R, Theta, Phi, 'j')
    Nr_m, Nt_m, Np_m = N_lm(l, -1, k, R, Theta, Phi, 'j')

    Mr_p, Mt_p, Mp_p = M_lm(l,  1, k, R, Theta, Phi, 'j')
    Mr_m, Mt_m, Mp_m = M_lm(l, -1, k, R, Theta, Phi, 'j')

    Er     += G_TM_p*Nr_p + G_TM_m*Nr_m
    Etheta += G_TM_p*Nt_p + G_TM_m*Nt_m + G_TE_p*Mt_p + G_TE_m*Mt_m
    Ephi   += G_TM_p*Np_p + G_TM_m*Np_m + G_TE_p*Mp_p + G_TE_m*Mp_m
    
    Hr     += (E0/z_imp)*(-G_TE_p*Nr_p - G_TE_m*Nr_m)
    Htheta += (E0/z_imp)*(-G_TE_p*Nt_p - G_TE_m*Nt_m + G_TM_p*Mt_p + G_TM_m*Mt_m)
    Hphi   += (E0/z_imp)*(-G_TE_p*Np_p - G_TE_m*Np_m + G_TM_p*Mp_p + G_TM_m*Mp_m)
```

Aplicamos un cambio de coordenadas y nos deshacemos de la dimensión extra en $\theta$ para poder representar el campo:

```python
Ex = (Er*np.sin(Theta)*np.cos(Phi) +
      Etheta*np.cos(Theta)*np.cos(Phi) -
      Ephi*np.sin(Phi))
    
Ey = (Er*np.sin(Theta)*np.sin(Phi) +
      Etheta*np.cos(Theta)*np.sin(Phi) +
      Ephi*np.cos(Phi))
    
Ez = Er*np.cos(Theta) - Etheta*np.sin(Theta)


Ex = np.squeeze(Ex)
Ey = np.squeeze(Ey)
Ez = np.squeeze(Ez)

E_plot = np.real(Ex)

Hx = (Hr*np.sin(Theta)*np.cos(Phi) +
      Htheta*np.cos(Theta)*np.cos(Phi) -
      Hphi*np.sin(Phi))
    
Hy = (Hr*np.sin(Theta)*np.sin(Phi) +
      Htheta*np.cos(Theta)*np.sin(Phi) +
      Hphi*np.cos(Phi))
    
Hz = Hr*np.cos(Theta) - Htheta*np.sin(Theta)


Hx = np.squeeze(Hx)
Hy = np.squeeze(Hy)
Hz = np.squeeze(Hz)

H_plot = np.real(Hy)
```

Representaciones gráficas

```python
R_mesh, P_mesh = np.meshgrid(r, theta, indexing='ij')

X = R_mesh * np.cos(P_mesh)
Y = R_mesh * np.sin(P_mesh)

plt.figure(figsize=(10, 8))
plt.pcolormesh(X, Y, E_plot, shading='auto', cmap='jet', vmax = 2, vmin = -2)
plt.colorbar(label='Re(Ex)')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Re(Ex) en Plano XZ (l_max={l_max})')
plt.axis('equal')
plt.show()
```

```python
R_mesh, P_mesh = np.meshgrid(r, theta, indexing='ij')

X = R_mesh * np.cos(P_mesh)
Y = R_mesh * np.sin(P_mesh)

plt.figure(figsize=(10, 8))
plt.pcolormesh(X, Y, H_plot, shading='auto', cmap='jet', vmax = 2, vmin = -2)
plt.colorbar(label='Re(Hy)')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Re(Hy) en Plano XZ (l_max={l_max})')
plt.axis('equal')
plt.show()
```
